```python
import zipfile
import pandas as pd

path = "../data/nhs_waiting_times.csv"

with zipfile.ZipFile(path, "r") as z:
    # find the first CSV inside
    csv_name = [n for n in z.namelist() if n.lower().endswith(".csv")][0]
    print("Loading:", csv_name)
    with z.open(csv_name) as f:
        df = pd.read_csv(f, encoding="latin1")

df.head()
```

```python
import pandas as pd

path = "../data/nhs_waiting_times.xlsx"

df = pd.read_excel(
    path,
    sheet_name="Table 1",
    header=2  # this tells pandas: "row 2 contains the column names"
)

df.head()
```

```python
# 1. Keep only regional rows
regions = df[df["Organisation Type"] == "Region"]

# 2. Select only the month columns (everything from Apr-25 onward)
month_columns = regions.columns[2:]

# 3. Add up missing submissions across all months for each region
regions["Total Missing Submissions"] = regions[month_columns].sum(axis=1)

# 4. Show regions ranked by total missing submissions
regions_sorted = regions[["Organisation Name", "Total Missing Submissions"]].sort_values(
    by="Total Missing Submissions",
    ascending=False
)

regions_sorted
```

```python
# 1. Keep only regional rows (make a proper copy)
regions = df[df["Organisation Type"] == "Region"].copy()

# 2. Select only the month columns
month_columns = regions.columns[2:]

# 3. Add up missing submissions across all months
regions["Total Missing Submissions"] = regions[month_columns].sum(axis=1)

# 4. Rank regions by total missing submissions
regions_sorted = regions[["Organisation Name", "Total Missing Submissions"]].sort_values(
    by="Total Missing Submissions",
    ascending=False
)

regions_sorted
```
